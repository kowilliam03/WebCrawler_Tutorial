### [Python讀寫檔](https://docs.python.org/3/library/functions.html#open)
```python
f = open('file name', 'mode')
```
- **file name**:要開啟的檔案檔名。
- **mode**:
    - ```r```:read, 檔案必須存在。(default)
    - ```w```:write，若檔案不存在，會新建一個檔案。若存在則會將原本內容清空。
    - ```a```:append，如果檔案在的話，會接在原先結尾後。  
    - ```t```：text mode(default)
    - ```b```: binary mode

#### 操作完成後，記得關閉檔案。 

````python
message = "Hello World"

f = open("test.txt", "w")
f.write(message)
f.close()
````

### 透過```with```，在操作完成後會自動關閉檔案。

````python
message = "Hello Python"
with open("test.txt", "a") as f:
    f.write(message)
````

## 實戰演練
### Dcard 寵物版

````python
import requests
from bs4 import BeautifulSoup
````

````python
url = "https://www.dcard.tw/service/api/v2/forums/pet/posts?limit=30"

resp = requests.get(url)
articles = resp.json()
article_id = [i['id'] for i in articles]
article_id
````

### 下載單一貼文中的圖片

````python
url = "https://www.dcard.tw/f/pet/p/{}".format(article_id[3])

resp = requests.get(url)
html = resp.text
html = BeautifulSoup(html, 'lxml')

imgs = html.find_all('img')
imgs
````

````python
imgs = html.find_all('img', class_="sc-1wz6w4a-0 XWDjG")
imgs
````

````python
def save_img(url, file_name):
    resp = requests.get(url)
    img = resp.content
    with open(file_name, "wb") as f:
        f.write(img)

for img in imgs:
    url = img.get('src')
    file_name = url.split("/")[-1]
    save_img(url, file_name)
````

### 用文章ID來分類圖片

````python
import os
````

````python
if not os.path.isdir('./img'):
    os.mkdir('img')
````

````python
def get_img_tag(id):
    url = "https://www.dcard.tw/f/pet/p/" + id
    resp = requests.get(url)
    
    html = resp.text
    html = BeautifulSoup(html, 'lxml')
    imgs = html.find_all('img', class_="sc-1wz6w4a-0 XWDjG") 
    
    return imgs

def save_img(url, dest):
    file_name = url.split("/")[-1]
    file_name = dest + '/' + file_name
    resp = requests.get(url)
    img = resp.content    

    with open(file_name, "wb") as f:
        f.write(img)
````

````python
dest = './img/'
for id in article_id:
    id = str(id) 
    tags = get_img_tag(id)
    
    for tag in tags:
        src = dest + id
        if not os.path.isdir(src):
            os.mkdir(src)
            
        url = tag.get('src')
        save_img(url, src)
````
